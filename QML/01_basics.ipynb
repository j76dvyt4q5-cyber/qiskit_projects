```python
import pennylane as qml
import numpy as np
import fastai
import fastbook
print("PennyLane version:", qml.__version__)
```

```python
dev = qml.device('default.qubit', wires=2)

@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0, 1])
    return qml.probs(wires=[0, 1])

print(circuit())
```

```python
dev = qml.device('default.qubit', wires=1)

@qml.qnode(dev)
def circuit(theta):
    qml.RY(theta, wires=0)
    return qml.probs(wires=0)

import numpy as np
for angle in [0, np.pi/4, np.pi/2, 3 * np.pi/4, 2*np.pi]:
    probs = circuit(angle)
    print(f"theta={angle:.2f} -> |0>={probs[0]:.2f}, |1>={probs[1]:.2f}")
```
